```python
import pandas as pd


trader_path = "historical_data.csv"
sentiment_path = "fear_greed_index.csv"

# Load CSVs
trader_df = pd.read_csv(trader_path)
sentiment_df = pd.read_csv(sentiment_path)


print("Trader columns:", trader_df.columns)
print("Sentiment columns:", sentiment_df.columns)
```

```python
# Standardize column names
trader_df.columns = trader_df.columns.str.strip().str.lower().str.replace(" ", "_")
sentiment_df.columns = sentiment_df.columns.str.strip().str.lower().str.replace(" ", "_")

trader_df.head(), sentiment_df.head()
```

```python
trader_df['parsed_time'] = pd.to_datetime(
    trader_df['timestamp_ist'],
    format="%d-%m-%Y %H:%M",
    errors='coerce'
)

trader_df['date'] = trader_df['parsed_time'].dt.date
sentiment_df['date'] = pd.to_datetime(
    sentiment_df['date'],
    dayfirst=True,
    errors='coerce'
).dt.date

print("Parsed sample trader times:\n", trader_df[['timestamp_ist','parsed_time']].head())
print("Sentiment sample:\n", sentiment_df[['date','classification']].head())
```

```python
agg_trader = trader_df.groupby('date').agg(
    total_pnl=('closed_pnl','sum'),
    total_trades=('account','count')
).reset_index()

print("Aggregated trader sample:\n", agg_trader.head())
```

```python
merged_df = pd.merge(
    agg_trader,
    sentiment_df[['date','classification']],
    on='date',
    how='inner'
)

print("Rows after merge:", len(merged_df))
print(merged_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(x='classification', y='total_pnl', data=merged_df, palette="viridis")
plt.title("Daily Total PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Daily Total PnL")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(
    x='classification',
    y='total_trades',
    data=merged_df.groupby('classification')['total_trades'].sum().reset_index(),
    palette="magma"
)
plt.title("Total Trades per Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Total Trades")
plt.show()
```

```python
avg_pnl = merged_df.groupby('classification')['total_pnl'].mean().reset_index()

plt.figure(figsize=(8,6))
sns.barplot(x='classification', y='total_pnl', data=avg_pnl, palette="coolwarm")
plt.title("Average Daily PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Daily PnL")
plt.show()

```

```python
merged_df['yesterday_sentiment'] = merged_df['classification'].shift(1)

plt.figure(figsize=(10,6))
sns.boxplot(x='yesterday_sentiment', y='total_pnl', data=merged_df, palette="coolwarm")
plt.title("Trader PnL vs Yesterday's Market Sentiment")
plt.xlabel("Yesterday's Sentiment")
plt.ylabel("Daily Total PnL")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(x='date', y='total_pnl', data=merged_df, marker="o", label="Daily Total PnL")

for sentiment in merged_df['classification'].unique():
    subset = merged_df[merged_df['classification'] == sentiment]
    plt.fill_between(subset['date'], subset['total_pnl'].min(), subset['total_pnl'].max(), alpha=0.1, label=f"{sentiment}")

plt.title("Daily PnL over Time with Sentiment Overlay")
plt.xlabel("Date")
plt.ylabel("Daily Total PnL")
plt.legend()
plt.show()
```
